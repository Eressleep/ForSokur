```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model  import LinearRegression
```

```python
train       = pd.read_csv('train.csv')
test        = pd.read_csv('test.csv')
feature     = ['x1', 'x2', 'x3', 'x4', 'x5']
test_x      = test[feature]
poly_test_x = PolynomialFeatures(degree=5).fit_transform(test_x)
train_x     = train[feature]
train_y     = train['y']

train
```

```python
poly_reg = PolynomialFeatures(degree=5)
x_poly   = poly_reg.fit_transform(train_x)

poly_reg = LinearRegression()
model    = poly_reg.fit(x_poly, train_y)
prediction = model.predict(poly_test_x)

prediction
```

```python
output = pd.DataFrame({'id': test.id,'y': prediction})
output.to_csv('submission.csv', index=False)
```
